## Policy Evaluation
* For prediction
    * input: $<S, A, P, R, \gamma>$
    * output: $v_\pi$
* For control
    * input: $<S, A, P, R, \gamma>$
    * output: $\pi_*$
* Iterative Policy Evaluation
    * problem: evaluate a policy $\pi$
    * sol: iterative application of Bellman expectation backup
    * update $v_{k+1}(s)$ from $v_k (s')$

## Policy Iteration
* $v_\pi (s) = E[R_{t+1} + \gamma R_{t+2} + ... | S_t=s]$
* $\pi' = greedy (v_\pi)$ 
* policy iteration always converges to $\pi_*$
