```python
%matplotlib inline
from sklearn import mixture
import numpy as np
import matplotlib.pylab as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
n_samples = 500
C = np.array([[0., -0.1], [1.7, .4]])
X = np.r_[np.dot(np.random.randn(n_samples, 2), C),
          .7 * np.random.randn(n_samples, 2) + np.array([-6, 3])]
print X.shape
```

```python
def getAICBIC(n_components):
    gmm = mixture.GMM(n_components=n_components, covariance_type='full').fit(X)
    return [gmm.aic(X), gmm.bic(X)]

components = list(range(1, 10))
res = np.array([getAICBIC(i) for i in components])
```

```python
plt.plot(components, res[:,0], color="red")
plt.plot(components, res[:,1], color="blue")
```
