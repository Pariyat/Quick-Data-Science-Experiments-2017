* Online Gradient Descent algorithm update rule - $w_{t+1} = w_t - \eta \hat{g_t}$
    * $\hat{g_t}$ is a subgradient of $f_t$, where $\hat{g_t} \in \frac{df_t(w_t)}{dw_t}$
    * where $f_t = l_t(w) + \lambda \| w \|$
* composite objective OGD - tries to solve the problem of the predictor (?? $w_t$) never going to 0
    * $w_{t+1} = \arg\min g_t w + \frac{1}{2 \eta} \|w - w_t \|^2$
* FTRL - $w_{t+1} = \arg \min_t ( g_{1:t} w  + t \lambda \|w\|_1 + r_{1:t} (w) )$
    * subgradient approxiamation + t copies of L1 penalty + stablization pentalty
    * results in a sparser solution because of t copies of the same penalty 

## FTRL - AdClick paper

* $g_t \in R^d$ are the X
    * $g_{1:t}$ is the sum of all vectors up to t 
* in logistic regression, $\nabla l_t (w) = (\sigma(w x_t) - y_t) x_t = (p_t - y_t) x_t$
* OGD isn't effective at producing sparse models
* $w_{t+1} = \arg \min_w (g_{1:t} w + \frac{1}{2} \sum_{s=1}^t \sigma_s \| w - w_s \|^2 + \lambda \|w\|  )$
    * $\sigma$ is a variable of the learning rate $\sigma_{1:t} = \frac{1}{\eta_t}$
    * when $\lambda=0$, it's the same as OGD
