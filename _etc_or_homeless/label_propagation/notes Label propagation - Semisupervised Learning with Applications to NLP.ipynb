### label propagation
* semi-supervised
* label propagation uses unlabeled instances during the process of finding out the labels

### graph model
* a graph where each node is an instance, each arc has a weight $T_{xy}$ which is prob of jumping from x to y
* an unlabeled node has soft labels (prob pos / neg)
* $T$ is a matrix holding all the weights of the graph
* $Y$ is a matrix holding all the soft labels e.g. $Y_{n1, r1}$ is the prob of n1 labeled as r1
