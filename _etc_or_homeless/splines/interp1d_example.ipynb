```python
%matplotlib inline
import numpy as np 
import scipy.interpolate as sp
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 10, 50)
yy = np.sin(xx)
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
```

```python
fl = sp.interp1d(x, y,kind='linear')
fc = sp.interp1d(x, y,kind='cubic')
```

```python
xnew = np.linspace(0, 10, 50)
plt.subplot(211)
# the real sin(x) function plot
plt.plot(xx, yy)
plt.legend(['sin(x)'], loc='best')
plt.subplot(212)
# the interpolation
plt.plot(x, y, 'o', xnew, fl(xnew), xnew, fc(xnew))
plt.legend(['sample', 'linear', 'cubic'], loc='lower left')
```
