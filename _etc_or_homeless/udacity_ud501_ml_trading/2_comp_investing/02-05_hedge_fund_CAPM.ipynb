### Two Stock Scenario
* Stock A, $B_A = 1.0$, predict +1% over market
* Stock B, $B_B = 2.0$, predict -1% over market
* Note: look at the slope of the two stocks and how they correspond to the beta

```python
Image("two_stock_scenario.png")
```

* since we assume that market doesn't move, see blue 0%, we can get rid of the $\beta_A r_m$ component and be left with $\alpha_A$
* also note that long means hold, short means sell
* so according to the market, we gain 0.5 dollar from both stocks

### Two Stock CAPM 
* recall $r_p = \sum_i w_i (\beta_i r_m + \alpha_i)$
* $= (w_A \beta_A + w_B \beta_B) r_m + (w_A \alpha_A + w_B \alpha_B)$
* $= (0.5 * 1 + -0.5 * 2) r_m + (0.5 * 0.01 + -0.5 * -0.01)$
* $= -0.5 r_m + 0.01$

### scenarios
* if $r_m = 0.1$, then $r_p = -0.5 * 0.1 + 0.01 = -0.04$
* if $r_m = -0.1$, then $r_p = -0.5 * -0.1 + 0.01 = 0.11$

* **note** if we can remove the $r_m$ term, then we can remove market risk
* therefore, we find $w_A \beta_A + w_B \beta_B = 0$
