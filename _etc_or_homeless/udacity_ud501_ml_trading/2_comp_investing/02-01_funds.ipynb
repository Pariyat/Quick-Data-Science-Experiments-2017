### Types of funds
#### ETF  (electronically traded funds)
* buy and sell like stocks 
* e.g. this ETF models the S&P 500
* basket of stocks 
* can do intraday
* **liquid**
* paid by expense ratio

#### Mutual Fund
* buy / sell at the end of day
* less transaprent
* **Large cap** - means # shares * stock
* paid by higher expense ratio 

#### Hedge Fund
* only buy / sell via agreement
* no disclosure
* **not transparent**
* paid by two and twenty, 2% Asset Under Management and 20% profit

### Two and Twenty
* $100,000,000 --> 15% profit (15M)

* 2% $100M + 20% * 15M = 2M + 3M = 5M

### How to attract investors
#### Who?
* individuals (e.g. individual with large $, because there's a cap on # of investors)
* institutions (e.g. Georgia Tech fund)
* fund of funds 

#### Why?
* track record
* simulation + story
* good portfolio fit (why ur strategy is complementary to their portfolio, e.g. if you only buy a specific niche)

### Hedge Fund Goals & Metrics
#### Goals
* beat a benchmark (S&P) (e.g. suck less than S&P)
* absolute return (Long / Short), small gradual returns

#### Metrics
* cumulative returns
* volatility (std of daily returns)
* risk / reward (sharpe ratio, or risk adjusted reward)
