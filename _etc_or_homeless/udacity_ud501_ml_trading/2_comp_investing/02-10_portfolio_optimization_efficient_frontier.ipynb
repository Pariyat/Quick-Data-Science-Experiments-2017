### importance of covariance
* say 3 stocks, ABC, DEF, GHI, all 10% increase
* ABC-DEF has 0.9 cov, ABC-GHI has -0.9 cov
* say a portfolio {50% ABC, 50% DEF}, return looks just like ABC or DEF
* say a portfolio {25% ABC, 25% DEF, 50% GHI}, return is a lot more smooth

### mean variance optimization (MVO)
* inputs
    * expected turn
    * volatility
    * covariance
    * target return (e.g. 5% return - **that black line below**)
* outputs 
    * asset weights for portfolio that minimizes risk (**orange dot!!**)
