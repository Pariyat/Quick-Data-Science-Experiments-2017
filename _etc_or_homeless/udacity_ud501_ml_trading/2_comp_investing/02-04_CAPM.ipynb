### definition of portfolio
* $\sum_i w_i = 1$
* however, you can also short a stock, so you can have negative w
* $\sum_i abs(w_i) = 1$
* return is $r_p(t) = \sum_i w_i r_i(t)$ **note: w_i can be neg**
* e.g. portfolio {0.75 stock A, -0.25 stock B, 1% stock A, -2% stock B}, return is 1.25%

### market portfolio
* cap weighted $w_i = \frac{mkt cap_i}{\sum_j mkt cap_j}$
* where mkt cap is num shares * price

### CAPM
* $r_i(t) = \beta_i r_m(t) + \alpha_i(t)$
* where $\beta_i r_m(t)$ asserts that a lot of the effect is due to the market
* $\alpha_i(t)$ is the residual, where $E(\alpha) = 0$
* $r$ is in terms of %

### CAPM vs Active Management
* passive: buy index and hold
* active: pick stocks (overweight some, underweight some)
* in terms of CAPM, both agree on $\beta_i r_i(t)$
* differ in that:
    * passive believe that $\alpha_i(t)$ is random, and can't predict
    * active believe that they can predict $\alpha_i$

### CAPM for portfolio
* $r_p(t) = \sum_i w_i (\beta_i r_m(t) + \alpha_i(t))$
* $= \beta_p r_m(t) + \sum_i w_i \alpha_i(t)$

### implications of CAPM
* $E(\alpha_i(t)) = 0$, then we can only cleverly choose $\beta_i$
* only way to beat market is to 
    * large $\beta$ when market is going up
    * low $\beta$ when market is down

### Arbitrage Pricing Theory
* break $\beta$ into different sectors (e.g. energy, finance)
* $r_i = \beta_{if} r_f + \beta_{ie} r_e + ... + \alpha_i$
