### what is Q
* Q[s,a] - immediate reward + discounted reward
* $\pi(s) = argmax_a Q[s,a]$, if you have already built the Q table

### Q Learning procedure
* experience tuple $<s, a, s', r>$

```python
Image("q_procedure.png")
```

### q-update 
* given experience tuple, improve Q table
* $\alpha$, $\gamma$ are [0,1]
* $Q[s,a] = (1 - \alpha) Q[s,a] + \alpha * new estimate$
* $= (1 - \alpha) Q[s,a] + \alpha (r + \gamma r_{later})$
* $= (1 - \alpha) Q[s,a] + \alpha (r + \gamma Q[s', argmax_a Q[s', a]])$ **note: recursive**

### randomness in action
* success depends on exploration
* choose random action w/ prob c (decays over iter, much like a bandit algo)

### state
* basically a single #
