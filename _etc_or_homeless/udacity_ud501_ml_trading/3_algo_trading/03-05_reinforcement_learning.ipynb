### RL problem
* environment state T that changes
* actor observes env s and rewards s
* actor updates $\pi(s)$
* actor takes action q
* find $\pi$

### trading to RL
* actions are {buy, sell, do nothing}
* states is basically the market and the features e.g. bollinger
* reward is the trades

### markov decision problem
* set of states S
* set of actions A
* transition function T[s, a, s'] - prob that from s, taking action a, we end up at s' - sums to 1
* reward function R[s, a] - gets a reward by taking action a from state s
