# Abstract

* evaluated influence of skew:
    * threshold metrics (Accuracy, F-score, Cohen’s kappa, and Krippendorf’s alpha)
    * rank metrics (area under the receiver operating characteristic (ROC) curve and precision-recall curve)
* every metric beside AUC are skewed for imbalanced datasets 
* recommend reporting skew-normalized scores along with the obtained ones

# Intro

* in imbalanced datasets, skew may spoil performance metrics
    * accuracy is vulnerable because of high # of negatives

# Data

* $skew = \frac{\text{neg examples}}{\text{pos examples}}$

# Methods

* $F_1 = 2* \frac{P*R}{P + R}$
* Cohen Kappa $K = \frac{P_{obs} + P_{chance} }{1 - P_{chance}}$
* Krippendorff’s α-reliability $\alpha = 1 - \frac{D_{obs}}{D_{chance}}$
