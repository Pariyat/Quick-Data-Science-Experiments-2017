#### MF algorithm:
* for row, col in R:
    * if not null: 
        * do SGD

* SVD $A_k = U_K \Sigma_k V_K^T$
* NMF $A_k = W_k H_k$
* columns of $W_k$ are the basis vectors, each col of $A_k$ can be built from cols of $W_k$
* $A_k e_1 = W_k H_{*1} = h_{11} W_{1} + h_{21} W_{2} + ... + h_{k1} W_k$

#### applications:

#### objective fn:
* MSE $min \|A - WH \|^2$ s.t. $W, H >= 0$
    * convex in W or H, not both, so hard to get to global min
* lots of other obj fns

#### ALS
* repeat steps below
* for H $W^T W H = W^T A$
    * should be reminicent of LS, $\beta = \frac{X^T Y}{X^T X}$ vs $H = \frac{W^T A}{W^T W}$
* for W $H^T H W = H^T A$
* only take H and W >= 0
* no sparsity set up into this
* no convergence theory
