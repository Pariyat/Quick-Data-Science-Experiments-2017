The bayes classification boundary is simply where the pdfs of two classes are equal.

### Consider the following
* class 1 ~ $(\mu_1, \Sigma_1)$ and class 2 ~ $(\mu_2, \Sigma_2)$
* \begin{equation} \mu_1 = \begin{pmatrix} -1 \\ -1 \end{pmatrix}, \Sigma_1 = \begin{pmatrix} 2 & 1/2 \\ 1/2 & 2 \end{pmatrix} \end{equation}
* \begin{equation} \mu_2 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}, \Sigma_2 = \begin{pmatrix} 1 & 0\\ 0 & 1 \end{pmatrix} \end{equation}
