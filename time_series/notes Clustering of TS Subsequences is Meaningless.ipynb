## Background

* common algorithms are hierarchical clustering and kMeans

## Meaninglessness of STS Clustering
* distance between 2 sets of clusters is $\text{cluster_dist}(A, B) = \sum_i^k min[ dist(\bar{a}, \bar{b} ]$
    * k is the number of clusters
* we can use this to measure to compare clusters from the same datasets or different datasets
* experiemnt
    * $\hat{x}$ - 3 sets of clusters on stock data
    * $\hat{y}$ - 3 sets of clusters on random walk
    * $\text{within_set_X_dist}$ - distance within set
    * $\text{between_set_X_Y_dist}$ - distance between set
    * $\text{cluster_meaningfulness}(\hat{x}, \hat{y}) = \frac{\text{within_set}}{\text{between_set}}$
    * should expect low number for meaningful clusters

### 4.1 hidden constraint
* when k=1, the cluster center is a straight line
* then the weighted avg of cluster centers when k > 1 must be a straight line as well

### 4.2 trivial match
* the closest match to $C_p$ is $C_{p-2}, C_{p-1}, C_p, C_{p+1}, C_{p+2}$

### 4.4 Necessary conditions for STS Clustering to work
* if we have a time series with k patterns of length w
* firstly, we need the k patterns to have a weighted sum of a straight line
* secondly, we need the same number of trivial matches for each pattern
