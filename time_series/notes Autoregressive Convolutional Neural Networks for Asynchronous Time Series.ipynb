* use conv net to learn non-linear relationships
* gating mechanism $f(x) = c(x) X \sigma(x)$, where X is an element wise matrix product
    * $\sigma(x) $ controls the amount of output passed to the next layer

### motivation
* combine simple AR w/ neural network to learn data dependent weights
* $E[x_n | {x_{n-m}, m=1..M}] = \sum_m \alpha_1 (x_{n-m}) * x_{n-m}$ - AR relationship
    * note: $\sum_m \alpha_m <= 1$
* propses to do $\alpha_m (x_{n-m}) * f(x_{n-m})$
* $\hat{y}_n^{i} = \sum_m [F(x_n^{-M}) X \sigma(S(x_n^{-M}))]_{im}$, where F and S are neural networks
* S is fully convolutional
* F is of the form $F(x_n^{-M}) = W X [off(x_{n-m}) + x_{n-m}^I]_{m=1}^M$
* called Significance offset  CNN (SOCNN), where offset is F, sig is S
