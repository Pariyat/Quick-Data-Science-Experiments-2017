## Granger Causality

### definition
* a cause is prior to its effect
* if X, Y have granger relationship, $X^{past}$ should help predict $Y^{future}$ than just using $Y^{past}$

### formal definition
* the causes happened prior to effect
* The cause makes unique changes in the effect. In other words, the causal series contains unique information about the effect series that is not available otherwise.
* let $I^*(t)$ be all info up to time t
* let $I^*_{-t}(x)$ be all the info excluding x
* granger causality $P(Y(t+1) \in I^*(x)) \neq P(Y(t+1) \in I^*_{-x}(x)) $
* math, basically perform 2 VARs
    * $Y(t) = \sum_{l=1}^L a_l Y(t-l) + \epsilon_t$
    * $Y(t) = \sum_{l=1}^L a^{'}_l Y(t-l) + \sum_{l=1}^L b^{'}_l X(t-l) + \epsilon_t$
    * X causes Y if the latter is statistically significantly better

### extension to multivariate time series
* given $X_1 ... X_p$
* exhaustive-granger - examine pairwise
* multivariate-granger - normal granger, but X contains all other Xs as well
